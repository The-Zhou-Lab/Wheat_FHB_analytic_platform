# Importing Libraries

```python
import cv2
import numpy as np
from skimage import measure, morphology
from skimage.morphology import binary_dilation
import matplotlib.pyplot as plt
```

# Red tags extracted

```python
def red_tages_extracted(image_path):
    # Read the image (BGR) and convert to HSV
    image = cv2.imread(image_path)
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define HSV range for red (note: red often requires two ranges, but here we use one as in your original code)
    lower_red = np.array([0, 120, 100])
    upper_red = np.array([10, 255, 255])
    mask = cv2.inRange(hsv, lower_red, upper_red)
    
    # Extract red regions using the mask
    red_extracted = cv2.bitwise_and(image, image, mask=mask)
    
    # Clean up the mask: label connected components and remove small objects
    labels = measure.label(mask)
    # Remove small objects (threshold based on image width)
    min_size = 0.01 * image.shape[1]
    cleaned_mask = morphology.remove_small_objects(labels, min_size=min_size)
    # Convert to binary and dilate to merge nearby regions
    cleaned_mask = cleaned_mask > 0
    selem = np.ones((10, 10), dtype=np.uint8)
    dilated_mask = binary_dilation(cleaned_mask, selem)
    
    # Label the dilated mask and extract bounding boxes for each connected component
    final_labels = measure.label(dilated_mask)
    bbox_set = []
    for region in measure.regionprops(final_labels):
        minr, minc, maxr, maxc = region.bbox  # bbox returns (min_row, min_col, max_row, max_col)
        bbox_set.append([minc, minr, maxc, maxr])  # Convert to (x1, y1, x2, y2)
    
    # Filter bounding boxes that are too close to each other (using a threshold of 2% of image width)
    threshold_distance = 0.02 * image.shape[1]
    bbox_red_sets = []
    for i, bbox1 in enumerate(bbox_set):
        add_bbox = True
        for bbox2 in bbox_set[i+1:]:
            distance = np.linalg.norm(np.array(bbox1[:2]) - np.array(bbox2[:2]))
            if distance < threshold_distance:
                add_bbox = False
                break
        if add_bbox:
            bbox_red_sets.append(bbox1)
    
    # Draw bounding boxes on a copy of the original image for display
    image_display = image.copy()
    for bbox in bbox_red_sets:
        x1, y1, x2, y2 = bbox
        cv2.rectangle(image_display, (x1, y1), (x2, y2), (0, 255, 0), 2)
    
    
    return bbox_red_sets,image_display

# Example usage:
if __name__ == "__main__":
    image_path="Data/2.jpg"
    bboxes,image_display = red_tages_extracted(image_path)
    image_display = cv2.cvtColor(image_display, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 10))
    plt.imshow(image_display)
    plt.axis("off")
    plt.title("Extracted Red Regions")
    plt.show()
    print("Red Tags Bounding Boxes:", bboxes)
```
